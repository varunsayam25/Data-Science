## Labelling

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
```

```python
SS=StandardScaler()
OHE=OneHotEncoder()
```

```python
df['Survived']=df['Survived'].astype(int)
df['Fare']=df['Fare'].astype(int)
df['Age']=df['Age'].astype(int)
```

```python
df['Age']=SS.fit_transform(df[['Age']])
df['Fare']=SS.fit_transform(df[['Fare']])
```

```python
df['Embarked']=OHE.fit_transform(df[['Embarked']]).toarray()
df['Sex']=OHE.fit_transform(df[['Sex']]).toarray()
```

```python
df.head()
```

## Trainning Data

```python
Target=df['Survived']
Features=df.drop('Survived',axis=1)
```

```python
Target
```

```python
Features
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import roc_auc_score,roc_curve,accuracy_score,confusion_matrix,classification_report
```

```python
x_train,x_test,y_train,y_test=train_test_split(Features,Target,train_size=0.75,random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## LGBM

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
Grad_boost= GradientBoostingClassifier(n_estimators=50,learning_rate=0.09,max_depth=3,subsample=0.4)
```

```python
Grad_boost.fit(x_train,y_train)
```

```python
y_predLGBM=Grad_boost.predict(x_test)
```

```python
accuracy_score(y_test,y_predLGBM)
```

```python
params={'n_estimators':[10,20,40,50,100],'learning_rate':[0.01,0.05,0.09],'max_depth':[1,2,3],'subsample':[0.1,0.2,0.3,0.4]}
```

```python
Grid_S=GridSearchCV(Grad_boost,params,verbose=2)
```

```python
Grid_S.fit(x_train,y_train)
```

```python
Grid_S.best_params_
```

## XGBM

```python
import xgboost as xgb
```

```python
xg_boost=xgb.XGBClassifier(n_estimators=50,learning_rate=0.05,max_depth=3,sub_sample=0.1,min_child_weight=2)
```

```python
xg_boost.fit(x_train,y_train)
```

```python
y_predXGB=xg_boost.predict(x_test)
```

```python
accuracy_score(y_test,y_predXGB)
```

```python
params1={'n_estimators':[10,20,40,50,100],'learning_rate':[0.01,0.05,0.09],'sub_samples':[0.1,0.2,0.3,0.4],'max_depth':[1,2,3],'min_child_weight':[1,2,3,4,5]}
```

```python
Grid_S=GridSearchCV(xg_boost,params1,verbose=2)
```

```python
Grid_S.fit(x_train,y_train)
```

```python
Grid_S.best_params_
```

## Comparision And Visulaization

```python
print(classification_report(y_test,y_predLGBM))
```

```python
print(classification_report(y_test,y_predXGB))
```

 Classification Report shows that XGBM has better results than LGBM

```python
LGBM_log_proba=Grad_boost.predict_log_proba(x_test)[:,1]
```

```python
auc_scoreLGBM=roc_auc_score(y_test,LGBM_log_proba)
auc_scoreLGBM
```

```python
fpr,tpr,thr=roc_curve(y_test,LGBM_log_proba)
```

```python
plt.plot(fpr,tpr,color='blue',lw=2,label=f'AUC={auc_scoreLGBM:.2}')
plt.plot([0,1],[0,1],linestyle='--',color='gray')
plt.xlabel('False Positive Rate(FPR)')
plt.ylabel('True Positive Rate(TPR)')
plt.title('ROC Curve')
plt.legend()
plt.grid(True)
```

```python
auc_scoreXGB=roc_auc_score(y_test,XGB_proba)
auc_scoreXGB
```

```python
fpr,tpr,thr=roc_curve(y_test,XGB_proba)
```
